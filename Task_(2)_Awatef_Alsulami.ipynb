## 1. Exploratory Data Analysis (EDA)

### Perform summary statistics on the dataset.

### Identify and analyze patterns in the data.

### Visualize the distribution of key variables (e.g., Traffic_Index, Date).

```python
sns.histplot(df['TrafficIndexLive'], bins=50, kde=True, color='blue')
plt.title('Distribution of Real-time Traffic Index')
plt.xlabel('Traffic Index')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
sns.histplot(df['Datetime'], bins=50, kde=True, color='blue')
plt.title('Datetime')
plt.xlabel('Datetime')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Explore relationships between variables (e.g., Traffic_Index vs. Weather_Condition).


```python
df11 = df.drop(columns=['TrafficIndexLive', 'TrafficIndexWeekAgo'])
corr = df11.corr(numeric_only=True)
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.violinplot( data=df ,orient='h' ,color='b' ,bw=1 ,cut=0)
plt.show()
```
